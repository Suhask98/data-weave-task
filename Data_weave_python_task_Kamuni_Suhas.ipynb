```python
""" 
Question 1
1. Number of URLH which are overlapping (Common) in two files.
"""
i=0 #initializing iterable variable
count = 0 # setting count to 0 so we can increment later
while i<len(y_data):
    if(y_data[i]['urlh']==t_data[i]['urlh']): #if urlh of ydata = today's urlh
        count+=1 # count increments
    i+=1
print('The number of overlapping URLH are {} in number '.format(count))
```

```python
"""
Question 2
For all the URLH which are overlapping, calculate the price difference (wrt available_price) if there is any between 
yesterday's and today's crawls (scraped data). There might be duplicate URLHs in which case you can choose the 
first valid (with http_status 200) record.
"""
i=0 #iterable variable
j=0 #iterable variable
count = 0
while i<len(y_data): #incrementing under yesterday's data
    while j<len(y_data): #incrementing under today's data
        if(y_data[i]['urlh']==t_data[j]['urlh']):
            if(y_data[i]['crawl_date']!=t_data[j]['crawl_date']): #consider only if crawl data is different 
                count=float(y_data[i]['available_price'])-float(t_data[j]['available_price']) # price difference 
        j+=1
    i+=1
print('The Price difference of is : {}'.format(count))
```

```python
"""
Question 3
Number of Unique categories in both files.
"""
i=0
count = 0
while i<len(y_data):
    if(y_data[i]['category']==t_data[i]['category']): #if the category matches
        count+=1
    i+=1
print('The number of unique categories :',len(y_data)-count) # printing the remainder,
#of total values and common values to find the number of unique categories
```

```python
"""
Question 4
Display List of categories which is not overlapping (Common) from two given files.
"""
common_category_list = [] #empty list to append later
i=0
while i<len(y_data):
    if(y_data[i]['category']==t_data[i]['category']): #condition for appending if the categories are common
        common_category_list.append(t_data[j]['category'])            
    i+=1        

def unique(list1):
    '''This function prints the unique values in a list without using predefined libraries'''
    # insert the list to the set
    list_set = set(list1)
    # convert the set to the list
    unique_list = (list(list_set))
    for x in unique_list:
        print("Unique category conserding both the files :",x)
        
unique(common_category_list)
```

```python
"""
Question 5
Generate the stats with count of urlh for all taxonomies (taxonomy is concatenation of category 
and subcategory separated by " > ") for today's file.

Eg: 
Cat1 > Subcat1: 3500 
Cat1 > Subcat2: 2000 
Cat2 > Subcat3: 8900 
"""
all_cat = {} #initialising an empty dictionary
for i in t_data: #incrementing over today's data
    if i['category'] not in all_cat:  # for appending category to our empty dictionary
        all_cat[i['category']]={}
    if i['subcategory'] not in all_cat[i['category']]: #for appending subcategory to our empty dictionary
        all_cat[i['category']][i['subcategory']]=set()
    all_cat[i['category']][i['subcategory']].add(i['urlh']) # for appending count of urlh to our empty dictionary
#printing our values according to the question
for cat,catval in all_cat.items():
    for subcat,subcatval in catval.items():
        print(cat,">",subcat,":",len(subcatval))
```

```python
"""
Question 6
Generate a new file where mrp is normalized. If there is a 0 or a non-float value or the key doesn't exist, make it "NA".
"""
y_norm = [] # initialising another empty list for yesterday's file
for i in open(r'C:\Users\ganga\Downloads\yesterday (1)\yesterday.json'):
    y_norm.append(json.loads(i)) # appending yday's data to the list
for i in y_norm:
    try: # using try to check for mrp float value
        i['mrp'] = float(i['mrp']) 
        if i['mrp'] == 0:
            i['mrp']= 'NA'#enter NA if mrp is 0
    except: # if already an NA exists, this function excludes it
        i['mrp']='NA'
f = open("demofile.json", "w") #opens an writable file
f.write(json.dumps(y_norm)) # appends the values
f.close() #closes the file
```

```python
"""
Question 7
Display the title and price of 10 items having least price.

Eg:
Title1 --> its price 
Title2 --> its price 
upto 10
"""
i=0
count_list = [] #initialising an empty list
for i in y_data:
    x=i['available_price'] #appening an x variable for available price of yesterday's data
    if(x==None): #if none value then continue in the loop
        continue
    else:
        x=float(x)
    count_list.append((i['title'],x)) #append the float values of the price
count_list.sort(key = lambda x: x[1])
print("Title \t\t\t  --> \t\t\t  price")
s=set()#empty list
for i in count_list:
    if(i[0] not in s):
        print(i[0],"-->",i[1])
        s.add(i[0])
    if(len(s)==10): # print the first 10 values
        break
```
